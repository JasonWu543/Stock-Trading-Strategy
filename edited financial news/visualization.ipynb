```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 读取 CSV 文件（假设文件在当前工作目录下或指定路径）
df = pd.read_csv('./merged_financial_news.csv')

# 2. 将 'date' 列转换为 datetime 类型
df['date'] = pd.to_datetime(df['date'])

# 3. 按天统计每一天的条目数
daily_counts = df.groupby(df['date'].dt.date).size()

# 4. 绘制折线图（如果数据量较大，折线图更易阅读）
plt.figure(figsize=(12, 6))
daily_counts.sort_index().plot(kind='line')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.title('Daily Article Counts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
